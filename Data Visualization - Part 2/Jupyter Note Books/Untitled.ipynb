```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('pokemon.csv')
df
```

```python
base_color = sb.color_palette()[0]
sb.countplot(data=df , y = 'type_1')
```

```python
sb.countplot(data=df , x = 'type_2')
```

```python
n_points = df.shape[0]
max_count = df['height'].value_counts()
print(max_count/n_points)
```

```python
sorted_counts = df['generation_id'].value_counts() 
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.4}); 
plt.axis('square') 
```
